```python
import pandas as pd
# load the dataset
superstore_df=pd.read_csv("../data/superstore.csv")

#to display first few rows
superstore_df.head()
```

```python
#get the number of rows and columns
superstore_df.shape
```

```python
#shows the columns names
print(superstore_df.columns)

#data types of columns
print(superstore_df.dtypes)
```

```python
#convert 'Order.Date' to datetime format
superstore_df['Order.Date']=pd.to_datetime(superstore_df['Order.Date'],errors='coerce')
print(superstore_df['Order.Date'].dtype)
```

```python
#drop meaningless columns
df=superstore_df.drop(columns=['记录数'])
df
```

```python
print(df.columns)
```

```python
#check missing values in each column
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#check unique values in each columns
for col in df.select_dtypes(include="object").columns:
    print(f"Column: {col}")
    print(df[col].unique()[:10])  # Displaying only the first 10 unique values
    print("-" * 50)
```

```python
#statistical summary
df.describe()
```

```python
#check for negative profit values
negative_profit_count=(df['Profit']<0).sum()
print(f"no of negative profit entries:{negative_profit_count}")
```

```python
#check which product or categories have most losses
negative_profit=df[df['Profit']<0]
negative_profit['Category'].value_counts()
```

```python
#region & markets with high losses
negative_profit['Region'].value_counts()
negative_profit['Market'].value_counts()
```

```python
negative_profit[['Discount','Profit']].describe()
```

```python
#convert Ship.date to datetime
df['Ship.Date']=pd.to_datetime(df['Ship.Date'])

#convert Row.ID to int
df['Row.ID']=df['Row.ID'].astype(int)

#check of data types again
df.info()
```
